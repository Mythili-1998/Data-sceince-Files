#####  OS Module

'''
os module is used for creating new directories, changing the directories, directories list, 
getting current working directories, rename the directories, remove the empty directories.
 
os. getcwd ():  The getcwd () function returns the current working directory.
os. chdir(path): The chdir () function changes the Current Working Directory.
os. mkdir (path [, mode]): The mkdir () function creates a new directory.
os. listdir(path): The function returns a list of names of all files and folders present in the Current   
                   Working Directory.
os. makedirs ():   This method helps to create multiple directories.  If the file already exists it throws File Exit Error.                         
os. removedirs ():  The function removes an existing directory including all sub-directories. It raises a FileNotFoundError if the directory is not found
                         .
 os. rmdir ():  The rmdir () function removes an existing directory. Unlike removedirs (), it does not 
                remove sub-directories.
 os. rename (old, new): The function renames a file or a directory. It needs two arguments, the 
                        current name of the file and the new name of the file.
os. Path. exists(path): It returns True if the passed file exists, otherwise, it returns False.

os.path.get_size (): In this method, python will give us the size of the file in bytes. To use this method, we need pass the name of the file as a parameter.
os. walk (): This method is used for getting the roots, directories and files.
'''

#### Sys Module 

'''
sys module in python is used to access variables & functions used by the interpreter.
Sys Module provides functions and variables used to manipulate the different parts of the python runtime environment.

Sys.argv:  Returns a list of command line arguments passed to a python script. The item at index 0 in this list is always the name of the script. The rest of the arguments are stored at the subsequent indices.we can access any command line arguments using sys.argv[1:]

Sys.path: This is an environment variable that is a search path for all Python modules.
Sys.exit:   This causes the script to exit back to either the Python console or the command prompt. This is generally used to safely exit from the program in case of generation of an exception.
Sys.version:  This attribute displays a string containing the version number of the current Python interpreter.
we can access any command line arguments using sys.argv[1:]
'''


```python
import sys
print(sys.version)
```

```python
age = 17
  
if age < 18: 
      
    # exits the program 
    sys.exit("Age less than 18")     
else: 
    print("Age is not less than 18")    # tells the python interpreter to quit
```

```python
# this code i already run in vs code
n = len(sys.argv)
print("total arguments passed:",n)
print("name of the file:",sys.argv[0])

for i in range(1,n):
    print(sys.argv[i],end="")

# Addition of numbers
sum = 0
for i in range(1,n):
    sum+=int(sys.argv[i])

print("total sum",sum)
```
